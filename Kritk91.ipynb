```python
#a
import sympy as sp
x,y = sp.symbols('x y')

f = (sp.exp(x))*(sp.sin(y)) + y**3

df_dx = sp.diff (f, x)

df_dy = sp.diff (f, y)

print("The first partial derivative of f with respect to x is", df_dx)
print("The first partial derivative of f with respect to y is", df_dy)
```

```python
#b gradient vector and magnitude 

#importing the necessary library
import sympy as sp

#defining function and variables
x, y = sp.symbols('x y')

f = (x**2)*y + (y**2)*x

#using sympy to find the gradient vector
gradient = sp.Matrix([sp.diff(f, number) for number in (x, y)])

#output
print("The gradient vector is", gradient)

point = [(x, 1), (y, -1)] #list of tuples 

#substiute values into the gradient vector 
gradient_at_point = gradient.subs(point)

#Compute the maginitude of the graident vector at the given point 
magnitude = sp.sqrt(sum(part_of_vector**2 for part_of_vector in gradient_at_point))
print("Magnitude of the gradient vector at the point is:", magnitude)
```

```python
#c second derivatives of (x^2)y + x(y^2)

#import necessary library
import sympy as sp

#defining function and variables
x, y = sp.symbols('x y')
f = (x**2)*y + x*(y**2)

#taking first partial derivatives
df_dx = sp.diff(f, x)

df_dy = sp.diff(f, y)

#second partial derivatives 
df_dxx = sp.diff(df_dx, x)
df_dxy = sp.diff(df_dx, y)

df_dyy = sp.diff(df_dy, y)
df_dyx = sp.diff(df_dy, x)

#output statements
print("The partial derivative of df_dx with respect to x is", df_dxx)
print("The partial derivative of df_dx with respect to y is", df_dxy)
print("The partial derivative of df_dy with respect to y is", df_dyy)
print("The partial derivative of df_dy with respect to x is", df_dyx)
```

```python
#d
#import the necessary libraries
import sympy as sp
import numpy as np 
import matplotlib.pyplot as plt

#define function using sympy
x, y = sp.symbols('x y')
j = x**3 - 3*x*y + y**3
#Convert the SymPy expression to a numerical function
j_func = sp.lambdify((x, y), j, 'numpy')

#Create a grid of x and y values and evaluate j on this grid
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

#Plot the contour
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
